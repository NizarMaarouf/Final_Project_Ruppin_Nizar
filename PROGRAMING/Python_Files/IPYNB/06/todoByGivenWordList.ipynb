```python
import csv
referencePath = (r'C:\Users\nizar.maarouf\Desktop\Final_Progect_Ruppin\DATA\CSV_Files\Data_Encoding.csv')
givenParagraph = (r'C:\Users\nizar.maarouf\Desktop\Final_Progect_Ruppin\DATA\TXT_FILES\Input_3.txt')

def WordToVector(givenWord):
    File = open(referencePath,'r',encoding="utf8")
    Reader = csv.reader(File)
    Data = list(Reader)
    returnValue = []
    for data  in Data:
        if((data) != ['', '', '', '', '', '', '', '', '', '']):
            if(data[0] == givenWord):
                returnValue = data
                break
         

    File.close()  
    
    return returnValue
    
def main(): 

    print()
    with open(givenParagraph, 'r',encoding="utf8") as file2:
        for element in file2:
            for word in element.split(' '):
                givenWord = word.strip()
                full_vector=WordToVector(givenWord)
                outcome = []

                if(full_vector != [] ):
                    #exist
                    #Filter vector
                    outcome = full_vector[len(full_vector) -1]
                    newVector = full_vector[1:len(full_vector)- 1]
                    print('Found - '+givenWord)
                    print(newVector)
                    print('outcome - ' + outcome)

                else:
                    # not exist
                    print('Wrong input - '+ givenWord)


                print('\n')

main()
```
