```python
def mean_absolute_difference(a):
    n = a.shape[0]
    total_gap = 0
    for i in range(n):
        x = a[i]
        for j in range(i):
            y = a[j]
            gap = np.abs(x - y)
            total_gap += gap
    n_pairs = n * (n - 1) / 2
    return total_gap / n_pairs

    mean_absolute_difference(5,6,3,2,8,7,1,5)
```
